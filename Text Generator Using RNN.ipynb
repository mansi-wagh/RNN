```python
# Import NumPy for numerical operations (arrays, indexing, etc.)
import numpy as np

# Import TensorFlow for building and training neural networks
import tensorflow as tf

# Sequential = simple stack of layers (one after another)
from tensorflow.keras.models import Sequential

# SimpleRNN = basic recurrent layer, Dense = fully connected output layer
from tensorflow.keras.layers import SimpleRNN, Dense
from tensorflow.keras import Input
```

```python
# -------------------- DATA PREPARATION --------------------

# Input text (this is what the model learns from)
text = "This is My First RNN"

# Get all unique characters from the text and sort them
chars = sorted(list(set(text)))

# Create dictionary: character в†’ index (number)
char_to_index = {char: i for i, char in enumerate(chars)}

# Create reverse dictionary: index в†’ character
index_to_char = {i: char for i, char in enumerate(chars)}
```

```python
# -------------------- CREATE TRAINING DATA --------------------

# Number of characters the model looks at to predict next one
seq_length = 3

# List to store input sequences
sequences = []

# List to store corresponding labels (next character)
labels = []

# Loop through text to create sequences
for i in range(len(text) - seq_length):

    # Take 3 characters as input
    seq = text[i:i + seq_length]

    # Take next character as output (label)
    label = text[i + seq_length]

    # Convert each character in sequence to its index
    sequences.append([char_to_index[char] for char in seq])

    # Convert label character to index
    labels.append(char_to_index[label])


# Convert lists into NumPy arrays
X = np.array(sequences)   # Input data
y = np.array(labels)      # Output labels
```

```python
# -------------------- ONE-HOT ENCODING --------------------

# Convert input into one-hot format
# Shape becomes: (samples, sequence_length, number_of_unique_chars)
X_one_hot = tf.one_hot(X, len(chars))

# Convert labels into one-hot format
y_one_hot = tf.one_hot(y, len(chars))
```

```python
# -------------------- BUILD MODEL --------------------
model = Sequential()

# Define input shape separately
model.add(Input(shape=(seq_length, len(chars))))
# Add RNN layer
# 50 = number of neurons (memory units)
# input_shape = (sequence length, number of features)
# Then add RNN layer (no input_shape here)
model.add(SimpleRNN(50, activation='relu'))

# Output layer
# len(chars) neurons because we predict one of many characters
# softmax gives probability distribution
model.add(Dense(len(chars), activation='softmax'))
```

```python
# -------------------- COMPILE MODEL --------------------

# optimizer='adam' в†’ adjusts weights efficiently
# loss='categorical_crossentropy' в†’ used for multi-class classification
# metrics=['accuracy'] в†’ track performance
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# -------------------- TRAIN MODEL --------------------

# Train model on data
# epochs=100 в†’ number of times model sees full dataset
model.fit(X_one_hot, y_one_hot, epochs=100)
```

```python
# -------------------- TEXT GENERATION --------------------

# Starting text to generate from
start_seq = "This is M"

# Store generated text (starts with initial string)
generated_text = start_seq

# Generate 50 new characters
for i in range(50):

    # Take last 3 characters from generated text
    x = np.array([[char_to_index[char] for char in generated_text[-seq_length:]]])

    # Convert to one-hot format
    x_one_hot = tf.one_hot(x, len(chars))

    # Predict probabilities for next character
    prediction = model.predict(x_one_hot, verbose=0)

    # Get index of highest probability
    next_index = np.argmax(prediction)

    # Convert index back to character
    next_char = index_to_char[next_index]

    # Add predicted character to text
    generated_text += next_char
```

```python
# Print final generated text
print("Generated Text:")
print(generated_text)
```
